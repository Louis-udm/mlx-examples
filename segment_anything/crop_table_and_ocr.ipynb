```python
from pdf2image import convert_from_path
import cv2
import numpy as np
import pytesseract
from tabulate import tabulate
import matplotlib.pyplot as plt


for i in range(4):

    # Read the image and convert to grayscale
    img = cv2.imread(f'/Users/zlu/Projects/llm-service/trying/out-{i}.png')
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    text = pytesseract.image_to_string(gray)
    print(text)
    print("=====================================")

    # method4
    # X Gradient 求x梯度
    xgrad = cv2.Sobel(gray, cv2.CV_16SC1, 1, 0)
    # Y Gradient #求y梯度
    ygrad = cv2.Sobel(gray, cv2.CV_16SC1, 0, 1)
    edges = cv2.Canny(gray, 90, 250, apertureSize=3)
    # cv2.imshow("Canny Edge", edges)
    # cv2.waitKey(0)
    
    tables = []
    lines = cv2.HoughLinesP(edges, 1, np.pi/180, 200, minLineLength=500, maxLineGap=10)
    for line in lines:
        print(type(line))
        x1, y1, x2, y2 = line[0]
        cv2.line(gray, (x1, y1), (x2, y2), (0, 0, 255), 2)
        # 将线的起点和终点添加到表格列表中
        if (x2-x1>50 and y2-y1<5) or (y2-y1>50 and x2-x1<5) :
            tables.append((x1, y1, x2, y2))

    # Display the table
    # cv2.imshow("line_detect_possible_demo", gray)
    # cv2.waitKey(0)

    x1=min([cood[0] for cood in tables])
    x2=max([cood[2] for cood in tables])
    y1=min([cood[1] for cood in tables])
    y2=max([cood[3] for cood in tables])
    plt.imshow(gray[y1:y2, x1:x2], cmap='gray')
    plt.show()


    # Use Tesseract to do OCR on the cropped table
    text = pytesseract.image_to_string(gray[y1:y2, x1:x2])
    print(text)


    # # Assuming the recognized text is tabular or can be split into a list of lists
    # rows = text.split("\n")
    # table_data = [row.split() for row in rows]

    # # Convert the table data to markdown
    # markdown = tabulate(table_data, tablefmt="pipe")

    # # Print the markdown
    # print(markdown)
```
